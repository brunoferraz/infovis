```python
df = statesRawData.query('date == "2020-05-24"')
```

```python
m = folium.Map(
    width = 600, height = 400,
    location=[-15.77972, -47.92972],
    zoom_start=3
)
m
```

```python
max_confirmed = statesRawData.max()[4]
min_confirmed = statesRawData.min()[4]
t = max_confirmed
bins = []
for i in range(0, 50):
    bins.append(int(t * i/50))
```

```python
colormap = cm.LinearColormap(colors=['#FFAA00','red'], index=[min_confirmed,max_confirmed],vmin=min_confirmed,vmax=max_confirmed)
colormap
```

```python
folium.Choropleth(
    geo_data = states_geo,
    name='choropleth',
    data=df,
    columns=["state","confirmed"],
    key_on='feature.properties.uf_05',
#     fill_color='OrRd',
    fill_color=colormap,
    fill_opacity=0.7,
    line_opacity=0.2,
    # bins = bins3,
).add_to(m)
m
```
